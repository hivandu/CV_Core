## gamma 变换， 伽马变换

```python
# 伽马变换
gamma_c = dog.copy()
rows = dog.shape[0]
cols = dog.shape[1]
deeps = dog.shape[2]

for i in range(rows):
    for j in range(cols):
        for d in range(deeps):
            gamma_c[i][j][d]= 255 * np.power(gamma_c[i][j][d]/255.0, 0.5)

# gamma_c = 255 * np.power(gamma_c/255.0, 0.5)
# gamma_c = gamma_c.astype('uint8')

print(gamma_c[100, 100, :])
print(gamma_c.shape)
print(dog.shape)

cv2.imwrite('assets/dog_img_gamma_c.jpg', cv2.hconcat([dog, gamma_c]))
# os.system('open assets/dog_img_gamma_c.jpg')
plt.imshow(trans2show(cv2.hconcat([dog, gamma_c])))
```

```python
gamma_c_20 = 255 * np.power(gamma_c/255.0, 0.2)
gamma_c_20 = gamma_c_20.astype('uint8')
gamma_c_80 = 255 * np.power(gamma_c/255.0, 0.8)
gamma_c_80 = gamma_c_80.astype('uint8')

plt.imshow(trans2show(cv2.hconcat([gamma_c_20, gamma_c_80])))
```

```python
# 灰度图伽马变换
gamma_c = dog_gray.copy()
rows = dog_gray.shape[0]
cols = dog_gray.shape[1]
for i in range(rows):
    for j in range(cols):
        gamma_c[i][j] = 255 * np.power(gamma_c[i][j]/255.0, 0.5)

cv2.imwrite('assets/dog_gray_gamma_c.jpg', cv2.hconcat([dog_gray, gamma_c]))
# os.system('open assets/dog_gray_gamma_c.jpg')
plt.imshow(trans2show(cv2.hconcat([dog_gray, gamma_c])))
```
